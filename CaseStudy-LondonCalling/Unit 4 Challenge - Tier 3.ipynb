**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

```python
bor_data_test = prop_bor_year.loc['Westminster']
```

```python
#Take indexed list of prices
def calc_price_ratio(bor_data):
    
    bor_data = bor_data.sort_index()
    #Calc ratio
    ratio = float(bor_data.iloc[-1]/bor_data.iloc[0])
    
    #Return ratio
    return ratio
```

## Finding numerical increase between 1995 and 2022

```python
#Take indexed list of prices
def calc_price_increase(bor_data):
    
    bor_data = bor_data.sort_index()
    #Calc difference
    diff = float(bor_data.iloc[-1] - bor_data.iloc[0])
    
    #Return difference
    return diff

boroughs = prop_bor_year.reset_index()['borough'].unique()

bordiffs = {}

for b in boroughs:
    # Pull year & price data for borough
    year_price = prop_bor_year.loc[b]
    
    bordiffs[b] = calc_price_increase(year_price)
    
#Turn dict to DataFrame, transpose, reset index
bordiffs_df = pd.DataFrame([bordiffs]).T.reset_index()

#Rename columns
bordiffs_df.rename({'index':'Borough', 0:'Increase'}, axis='columns', inplace=True)

#Sort by ratio
bordiffs_sorted = bordiffs_df.sort_values('Increase', ascending=False)

bordiffs_sorted.reset_index()
```

```python
#bordiffs_df.plot(x='Borough', y='Increase', kind='bar')
bordiffs_df.plot(x='Borough', y='Increase', kind='bar', figsize=(15,5), title='Increase in housing price in Pounds', legend=False, ylabel='Millon £')
```
